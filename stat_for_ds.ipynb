```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt

%matplotlib inline
```

# **MEASURE OF CENTRAL TENDENCY**

```python
df = pd.DataFrame({"movie":['a','b','c','a','b','c','a','b','c'], "rating": [9,7,6,8,9,7,6,8,5]})
```

```python
df.mean()
```

```python
sns.displot(df,kde = True,rug =True)
plt.axvline(np.mean(df.rating), color ='r',linestyle ='-')
plt.axvline(np.median(df.rating), color ='g',linestyle ='-')
plt.axvline((df.rating.mode().values[0]), color ='y',linestyle ='-')
```

**Varience and standard deviation **

```python
df.rating.var()
```

```python
df.rating.std()
```

```python
mean = df.groupby(['movie'])['rating'].mean()
std = df.groupby(['movie'])['rating'].std()

mean
```

```python
std
```

```python
fig,ax = plt.subplots()
mean.plot.bar(yerr = std ,capsize = 4)
```

***Quartile and Inter Quartile***

```python
df2 = sns.load_dataset('tips')
sns.set_theme(style = 'whitegrid')
ax = sns.boxplot(x='day',y='total_bill',data = df2)
```

```python
ax = sns.boxplot(x='day',y='total_bill',data = df2)
ax = sns.swarmplot(x='day',y='total_bill',data = df2, color ='0.25')
```

```python
print(df2['total_bill'].quantile([0.05,0.25,0.5,0.75]))
```

***INTER QUARTILE RANGE***

```python
print(df2['total_bill'].quantile(0.75) - df2['total_bill'].quantile(0.25))
```

# CORRELATION AND COVARIENCE

```python
fig, ax = plt.subplots(figsize=(6,6))
ax = sns.heatmap(df2.corr(),vmin = -1 ,vmax = 1, cmap = sns.diverging_palette(20,220,as_cmap =True),ax=ax)
plt.tight_layout()
plt.show()
```
